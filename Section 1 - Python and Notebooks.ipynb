# Reading Datasets
Let us now import a library called *Pandas*, which is very powerful for data management, resampling, and plotting.

```python
import pandas as pd
# here, 'pd' is an alias for the imported Pandas library
```

Let's use the Pandas library to open some data on the Pamba Basin stored in an Excel file. In the current directory there is a 'Data' folder which contains this information. We will assign this data to a variable called `Dat`, which is a *DataFrame*, and then look at the first five rows of data using the `head()` method.

```python
Dat=pd.read_excel('./Data/Discharge_Pambabasin.xlsx')
Dat.head()
```

Note that the command reads the first sheet in the Excel file (by default unless otherwise specified). It also reads in all the columns and then automatically assigns the column names as the first row of the Excel sheet. The last six columns of this worksheet are blank, so we could a) specify the specific sheet we want and b) import just the columns we want using the following.

```python
Dat=pd.read_excel('./Data/Discharge_Pambabasin.xlsx' , sheet_name="KURUDAMANNIL" , usecols=[0,1,2,3,4,5,6])
Dat.head()
```

The first integers on the left is a generic sequential index. But we also recognize that the first labeled column is a date, and the third column is the time of observation - let's create a date-time index, rather than an ordinal index.

```python
Dat=pd.read_excel('./Data/Discharge_Pambabasin.xlsx' , sheet_name="KURUDAMANNIL" , usecols=[0,1,2,3,4,5,6] , \
                  parse_dates=[['Date','Time']], index_col=0)
Dat.head()
```

# Plotting and Time-Series Resampling
Now that we now have a *Data Frame* named $Dat$, and we have the index assigned as real date-time value, we can now do a lot of powerful things with Pandas. For example, we can make a simple plot of *Discharge* by selecting the data we want using `Dat.Discharge` and using the `.plot()` *method*. 

A *method* is a function operating on the object, in this case, a *Data Frame*.

```python
Dat.Discharge.plot()
```

Using the date-time index, we can also slice, modify, and select different data. For example, we can look at just data from 2010.

```python
Dat[Dat.index.year==2010].Discharge.plot()
```

Or we can look at the average flow by year by using the `groupby` method.

```python
Dat.groupby(Dat.index.year).mean().Discharge.plot()
```

Make a plot of the *maximum* monthly flow, but this time using the `resample` method and specifying to resample on a monthly basis `M`.

```python
Dat.resample('M').max().Discharge.plot()
```

We can also bring several plots together into one, and then start to add *aesthetics* with color.

```python
Dat.resample('M').min().Discharge.plot(c='gray')
Dat.resample('M').mean().Discharge.plot(c='blue')
Dat.resample('M').max().Discharge.plot(c='red')
```

# Customizing Plots
We can also:

- make the plot a little bigger by specifying the `figsize`
- change the yscale to `log`
- add a plot `title`
- add a `grid`
- add a `ylabel`
- add a `legend` and specify the labels

```python
plt.figure(figsize=(9,6)) # specify the figure size here.
Dat.resample('M').min().Discharge.plot(c='gray')
Dat.resample('M').mean().Discharge.plot(c='blue')
Dat.resample('M').max().Discharge.plot(c='red')
plt.yscale('log') # change the y-scale to log
plt.title('Flows at Kurudamannil') # add a plot title
plt.grid() # this adds a grid to major intervals
plt.ylabel('Discharge') # add a label to the y-axis
plt.legend(['Monthly min' , 'Monthly mean' , 'Monthly max']) # create custom legend labels for plotted data
```

# Scatter Plots
We can also make scatters of the data; in this case, a scatter of the *Gauge* and *Discharge* values will represent a rating curve of the observation site. This plot can indicate:

1. Erroneous gage readings
1. Mobile bed affecting the stage-discharge relationship
1. Hysteresis

```python
Dat.plot('Discharge','Gauge',kind='scatter') ; plt.grid()
```

So what is happening over time with this gage? Is this hysteresis, or is the rating curve changing with time? Let's change the color of each dot using the year of the sample and see if there's any patterns. Pandas is smart enough to conver the range into a color scale, which is specified as the year using `c=Dat.index.year`. We specify special color map using `cmap=viridis`. Change of the x-scale to log helps to differentiate between the years.

```python
Dat.plot('Discharge' , 'Gauge' , kind='scatter' , c=Dat.index.year , colormap='viridis')
# Other CMAP Options: 'viridis', 'plasma', 'inferno', 'magma', 'cividis'
plt.xscale('log')
plt.xlabel('Discharge')
```

# Rolling Means - Estimation of 7Q10
We can also apply *rolling* methods, which are moving windows with some method applied to the window. This may be useful in calcuation of the 7Q10, for example, or the 10% non-exceedance 7-day moving average. Since we have three measurements a day, let's find the rolling mean for 21 periods, or `7*3`, and assign to a new parameter called `Q7R`.

```python
Q7R = Dat.Discharge.rolling(7*3).min()
Q7R.plot()
```

`Q7R` still has the date-time index, therefore we can simply ask for the annual minimum value and assign to `AnnMins`.

```python
AnnMins = Q7R.resample('A').min()  #  here, 'A' is for Annual
AnnMins
```

In general, 15 years of data is insufficient to estimate a 10% non-exceedance level, but the process is demonstrated here. Now, with a list of the annual minimum rolling 7-day values, we can find the 10% level.

```python
AnnMins.quantile(.1)
```

# Statistical analyses - a trend test
We can also do some statistical analysis using the `scipy.stats` library. We've seen what appears to be some trends in the flow data, but let's evaluate closer using the non-parametric trend test, *Mann-Kendall*. Let's start by importing the function we want, `kendalltau`, resampling data for only specific months, and then doing a trend test.

```python
from scipy.stats import kendalltau
```

Here, we group by the index Year and Month, finding the minimum. Note that the returned DataFrame has a *multi-index*, or is considered hierarchical. The first column of the index is the year, next is the month.

```python
MonthQ = Dat.groupby([Dat.index.year.values , Dat.index.month.values]).min()
MonthQ.head(14) #look at the first 14 rows
```

We can extract a specific month and then plot; in this case, we're looking for trends in the monthly minimum flow over the period of record.

```python
MonthQ5=MonthQ.loc[(slice(None) , 5) , :].Discharge # we are getting the 5th month for all years
MonthQ5.plot()
```

Now we can do a trend-test on the above monthly minimum flows. The p-value is less than 5% significance indicating that this is a statistically significant positive trend, confirming what appears visually above.

```python
kendalltau(range(len(MonthQ5)) , MonthQ5)
```
